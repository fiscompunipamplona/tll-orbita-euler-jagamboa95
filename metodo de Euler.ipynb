```python
from numpy import loadtxt,array,arange,linspace,sin,cos
from pylab import plot, show, savefig, ylim, xlim, grid, xlabel, ylabel,imshow
import matplotlib.pyplot as plt
```

```python
m=1
r0=1
v0=2*3.1416
dt=0.02
n=200
to=0
theta0=0
GM=(4*(3.1416)**2)/(r0)**2 # UA^3 /yr^2
r=[]
v=[]
t=[]
x=[]
y=[]
e=[]
theta=[]
for i in range(n):
        
    ri=r0+(v0*dt)
    vi=v0+GM*dt
    r0=ri
    v0=vi
    to=to+dt
    x0= r0 * sin(theta)
    y0= r0 * cos(theta)
    theta0= theta0 + (vi/ri)*dt
    theta.append(theta0)
    r.append(r0)
    v.append(v0)
    t.append(to)
    x.append(x0)
    y.append(y0)
    
    #print(to,  r0)

plt.plot(t,r,'b')
grid()
show()

plt.plot(t,v,'g')
grid()
show()

plt.polar(theta,r,'r')
grid("on")
show()
```
